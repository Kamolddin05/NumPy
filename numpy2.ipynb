```python
"""                   Array Concatenation and Splitting                                 """
```

```python
# ----- Concatenation of arrays ----- #
```

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
z = [99, 99, 99]
print(np.concatenate([x, y, z]))
```

```python
grid = np.array([[1, 2, 3],
                [4, 5, 6]])
np.concatenate([grid, grid]) # gorizontal birlashtirish
```

```python
grid = np.array([[1, 2, 3, 4],
                 [5, 6, 7, 8]])
np.concatenate([grid, grid], axis= 1) # vertikal birlashtirish
```

```python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7],
               [6, 5, 4]])
np.vstack([x, grid])
```

```python
y = np.array([[99],
              [99]])
np.hstack([grid, y])
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8,])
z = np.array([9, 10, 11, 12])
np.concatenate([x, y, z], axis= 0) # gorizontal
```
