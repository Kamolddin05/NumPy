```python
# Boolean operators
```

```python
np.sum((inches > 0.5) & (inches < 1))
```

```python
inches > (0.5 & inches) < 1

# Operatorlarning ustuvorligi sababli qavslarni olib tashlasak, ifoda noto‘g‘ri talqin qilinadi:
# chunki Python operatorlarni noto‘g‘ri tartibda bajaradi.
```

```python
np.sum(~((inches <= 0.5) | (inches >= 1)))
```

```python
print("Number days without rain: ", np.sum(inches == 0))
print("Number days with rain: ", np.sum(inches != 0))
print("Days with more than 0.5 inches:", np.sum(inches > 0.5))
print("Rainy days with < 0.1 inches :", np.sum((inches > 0) &
                                              (inches < 0.2)))
```

```python
# ----- Boolean Arrays as Masks ----- #
```

```python
x = np.array([[5, 0, 3, 3],
              [7, 9, 3, 5],
              [2, 4, 7, 6]])
x
```

```python
x < 5
```

```python
x[x < 5]
```

```python
rainy = (inches > 0)

summer = (np.arange(365) - 172 < 90) & (np.arange(365) - 172 > 0)
print("Median precip on rainy days: ", np.median(inches[rainy]))
print("Median precip on summer days:", np.median(inches[summer]))
print("Maximum precip on summer days:", np.max(inches[summer]))
print("Median precip on non-summer rainy days:",
      np.median(inches[rainy & ~summer]))
```
